```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x=np.random.rand(100,1) - 0.5
y=3*x[:,0]**2 + 0.05*np.random.randn(100)
```

```python
df=pd.DataFrame()
```

```python
df["X"]=x.reshape(100)
df["Y"]= y
```

```python
df
```

```python
plt.scatter(df["X"],df["Y"])
plt.title(" X vs Y")
```

```python
df["pred1"]=df["Y"].mean()
```

```python
df
```

```python
df["red1"]=df["Y"]-df["pred1"]
```

```python
df
```

```python
plt.scatter(df["X"],df["Y"])
plt.plot(df["X"],df["pred1"],color="red")
plt.title(" X vs Y")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt1=DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
dt1.fit(df["X"].values.reshape(100,1),df["red1"])
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dt1)
plt.show()
```

```python
x_test=np.linspace(-0.5,0.5,500)
y_pred=0.265458 + dt1.predict(x_test.reshape(500,1))
```

```python
plt.plot(x_test,y_pred,linewidth=2, color="red")
plt.scatter(df["X"],df["Y"])
```

```python
df["red2"]=df["Y"]-df["pred2"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

def gradient_boost(x, y, lr, number, count=1, regs=None, foo=None):
    if regs is None:
        regs = []
    
    if number == 0:
        return
    
    # Compute residuals
    if count > 1:
        y = y - lr * regs[-1].predict(x)
    else:
        foo = y
    
    tree_regs = DecisionTreeRegressor(max_depth=5)
    tree_regs.fit(x, y)
    regs.append(tree_regs)
    
    x_test = np.linspace(-0.5, 0.5, 500).reshape(-1, 1)
    y_pred = sum(lr * regressor.predict(x_test) for regressor in regs)
    
    print(f"Iteration {count}")
    plt.figure()
    plt.plot(x_test, y_pred, linewidth=0.8, color="red", label="Boosted prediction")
    plt.scatter(x[:, 0], foo, color="blue", s=10, label="True target")
    plt.legend()
    plt.show()
    
    gradient_boost(x, y, lr, number-1, count+1, regs=regs, foo=foo)


np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)

gradient_boost(X, y, lr=0.5, number=5)
```
