```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_regression
x,y=make_regression(n_samples=100,n_features=1,n_targets=1,n_informative=1,noise=20,random_state=20)
```

```python
plt.plot(x,y,"b.")
plt.show()
```

```python
class meraRidge:
    def __init__(self,alpha=0.001):
        self.alpha=alpha
        self.intercept_=None
        self.coef_=None

    def fit(self,x_train,y_train):
        num=0
        dem=0
        for i in range(x_train.shape[0]):
            num= num +((y_train[i]-y_train.mean())*(x_train[i]-x_train.mean()))
            dem= dem + ((x_train[i]-x_train.mean())*(x_train[i]-x_train.mean()))

        self.coef_= num/(dem + self.alpha)
        self.intercept_ = y_train.mean() - self.coef_*x_train.mean()
        print(self.intercept_,self.coef_)

    def predict(self,x_test):
        return self.coef_*x_test + self.intercept_
```

```python
MR=meraRidge(10)
```

```python
MR.fit(x,y)
```

```python
MR.predict(0.334)
```
