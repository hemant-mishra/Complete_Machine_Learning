```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False, class_sep=20)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap="winter",s=100)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(penalty=None,solver="sag")
LR.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
from sklearn.metrics import recall_score, precision_score
```

```python
y_pred=LR.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
print(recall_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
```
