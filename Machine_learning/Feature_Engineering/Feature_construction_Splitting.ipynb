```python
#feature construction
```

```python
df["family_size"]=df["SibSp"]+df["Parch"]+1
```

```python
df
```

```python
def Ftype(num):
    if num==0:
        return 0
    elif num>1 and num<4:
        return 1
    else:
        return 2
```

```python
df["Family_type"]=df["family_size"].apply(Ftype)
```

```python
df.drop(columns=["SibSp","Parch","family_size"],axis=1,inplace=True)
```

```python
df
```

```python
print(np.mean(cross_val_score(LogisticRegression(),df[["Age","Pclass","Family_type"]],df["Survived"],scoring="accuracy",cv=20)))
```

```python
df=pd.read_csv(r"C:\Users\heman\Downloads\train.csv")
```

```python
df=df[["Age",'Pclass',"SibSp","Parch","Survived","Name"]]
```

```python
df["Title"]=df["Name"].str.split(",",expand=True)[1].str.split(".",expand=True)[0]
```

```python
df
```

```python
df.drop(columns=["Name"],axis=1,inplace=True)
```

```python
df.groupby("Title").mean()["Survived"]
```

```python
df["is_married"]=0
```

```python
df['Is_Married'] = 0
df['Is_Married'].loc[df['Title'] == 'Mrs'] = 1
```

```python
def married(value):
    if value=="Mrs":
        return 1
    else:
        return 0
```

```python
df["Is_Married"]=df["Title"].apply(married)
```

```python
df.drop(columns=["is_married"],axis=1,inplace=True)
```
