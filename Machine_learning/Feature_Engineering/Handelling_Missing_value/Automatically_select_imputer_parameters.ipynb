```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
numerical_features=["Age","Fare"]
numerical_tranforemer=Pipeline(steps=[("SI",SimpleImputer(strategy="median")),("SS",StandardScaler())])

categorical_feature=["Sex","Embarked"]
Categorical_transformer=Pipeline(steps=[("SIC",SimpleImputer(strategy="most_frequent")),("OHE",OneHotEncoder(handle_unknown="ignore"))])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
trf=ColumnTransformer(transformers=[
    ("Num_trans",numerical_tranforemer,numerical_features),
    ("cat_trans",Categorical_transformer,categorical_feature)
])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Clf=Pipeline(steps=[
    ("trf",trf),
    ("Model",LogisticRegression())
])
```

```python
from sklearn import set_config
set_config(display="diagram")
Clf
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={"pre_num_imputer":["mean","median"],"pre_cat_imputer":["most_frequent","constant"],"classifier__C":[0.1,1,10,100]}
grid_search=GridSearchCV(Clf,param_grid,cv=10)
```

```python
grid_search.fit(x_train,y_train)
```
