```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv(r"C:\Users\heman\Downloads\train.csv",usecols=["Age","Pclass","Fare","Survived"])
```

```python
df
```

```python
df.isnull().mean()*100
```

```python
x=df[["Age","Fare","Pclass"]]
y=df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.impute import KNNImputer, SimpleImputer
```

```python
knn=KNNImputer(n_neighbors=5,weights="distance")
x_train_transform=knn.fit_transform(x_train)
x_test_transform=knn.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
LG=LogisticRegression()
LG.fit(x_train_transform,y_train)
y_pred=LG.predict(x_test_transform)
print(accuracy_score(y_test,y_pred))
```

```python
# comparision with simple imputer 
```

```python
SI=SimpleImputer()
x_tarin_trf=SI.fit_transform(x_train)
x_test_trf=SI.transform(x_test)
```

```python
LGS=LogisticRegression()
LGS.fit(x_tarin_trf,y_train)
y_pred1=LGS.predict(x_test_trf)
print(accuracy_score(y_test,y_pred1))
```
