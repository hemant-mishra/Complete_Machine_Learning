```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train_Age_mean=x_train["Age"].mean()
x_test_Age_mean=x_test["Age"].mean()
x_train_Fare_median=x_train["Fare"].median()
x_test_Fare_median=x_test["Fare"].median()
```

```python
x_train["Age_mean"]=x_train["Age"].fillna(x_train_Age_mean)
x_test["Age_mean"]=x_test["Age"].fillna(x_test_Age_mean)
x_train["Fare_median"]=x_train["Fare"].fillna(x_train_Fare_median)
x_test["Fare_median"]=x_test["Fare"].fillna(x_test_Fare_median)
```

```python
x_train_Age_median=x_train["Age"].median()
x_test_Age_median=x_test["Age"].median()
x_train_Fare_mean=x_train["Fare"].mean()
x_test_Fare_mean=x_test["Fare"].mean()
```

```python
x_train["Age_median"]=x_train["Age"].fillna(x_train_Age_median)
x_test["Age_median"]=x_test["Age"].fillna(x_test_Age_median)
x_train["Fare_mean"]=x_train["Fare"].fillna(x_train_Fare_mean)
x_test["Fare_mean"]=x_test["Fare"].fillna(x_test_Fare_mean)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
imputer1=SimpleImputer(strategy="mean")
imputer2=SimpleImputer(strategy="median")
```

```python
trf=ColumnTransformer(transformers=[
    ("impt1",imputer1,["Age"]),
    ("impt2",imputer2,["Fare"])
],remainder='passthrough')
```

```python
trf.fit(x_train)
```

```python
x_train_transform=trf.transform(x_train)
x_test_transform=trf.transform(x_test)
```

```python
x_train["Age_99"]=x_train["Age"].fillna(99)
x_train["Age_1"]=x_train["Age"].fillna(-1)
x_train["Fare_999"]=x_train["Fare"].fillna(999)
x_train["Fare_1"]=x_train["Fare"].fillna(-1)
```
