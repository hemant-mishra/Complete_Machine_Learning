```python
import pickle
import numpy as np
```

```python
clf=pickle.load(open("titanic_model.pkl","rb"))
ohe_sex=pickle.load(open("OHE_sex.pkl","rb"))
ohe_embarked=pickle.load(open("OHE_Embarked.pkl","rb"))
```

```python
test_input=np.array([2,"male",32,40,0,"S"],dtype=object).reshape(1,6)
```

```python
test_input
```

```python
test_input_sex=ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_sex
```

```python
test_input_Embarked=ohe_embarked.transform(test_input[:,-1].reshape(1,1))
```

```python
test_input_Embarked
```

```python
test_input_transformed=np.concatenate((test_input[:,[0,3,4]],test_input[:,2].reshape(1,1),test_input_sex,test_input_Embarked),axis=1)
```

```python
test_input_transformed
```

```python
clf.predict(test_input_transformed)
```
