```python
dependent_columns=df1["Survived"]
```

```python
independent_columns=df1.iloc[:,1:]
```

```python
independent_columns.sample(5)
```

```python
dependent_columns.sample(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent_columns,dependent_columns,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
from sklearn.impute import SimpleImputer
SI=SimpleImputer(strategy="most_frequent")

x_train_age=pd.DataFrame(SI.fit_transform(x_train[["Age"]]),columns=["Age"])
x_train_embarked=pd.DataFrame(SI.fit_transform(x_train[["Embarked"]]),columns=["Age"])

x_test_age=pd.DataFrame(SI.fit_transform(x_test[["Age"]]),columns=["Age"])
x_test_embarked=pd.DataFrame(SI.fit_transform(x_test[["Embarked"]]),columns=["Age"])
```

```python
from sklearn.preprocessing import OneHotEncoder

OHE_sex=OneHotEncoder(drop="first",sparse=False)
x_train_sex=OHE_sex.fit_transform(x_train[["Sex"]])
x_test_sex=OHE_sex.fit_transform(x_test[["Sex"]])
```

```python
x_train_sex.shape
```

```python
x_train["Pclass"].values.reshape(623,1)
```

```python
from sklearn.preprocessing import OneHotEncoder

OHE_Embarked=OneHotEncoder(drop="first",sparse=False)
x_train_Embarked_ohe=OHE_Embarked.fit_transform(x_train_embarked)
x_test_Embarked_ohe=OHE_Embarked.fit_transform(x_test_embarked)
```

```python
x_train_Embarked_ohe
```

```python
x_train_final=np.hstack((x_train["Pclass"].values.reshape(623,1),x_train_sex,x_train_age,x_train["Fare"].values.reshape(623,1),x_train_Embarked_ohe,x_train["family"].values.reshape(623,1)))
```

```python
x_train_final[1]
```

```python
x_test_final=np.hstack((x_test["Pclass"].values.reshape(268,1),x_test_sex,x_test_age,x_test["Fare"].values.reshape(268,1),x_test_Embarked_ohe,x_test["family"].values.reshape(268,1)))
```

```python
x_test_final[1]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train_final,y_train)
```

```python
y_pred=clf.predict(x_test_final)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_sc=accuracy_score(y_test,y_pred)
```

```python
print(acc_sc)
```
