```python
df["cgpa_z"]=(df["cgpa"]-df["cgpa"].mean())/df["cgpa"].std()
```

```python
lower_limit=df["cgpa"].mean()-3*df["cgpa"].std()
```

```python
uper_limit=df["cgpa"].mean()+3*df["cgpa"].std()
```

```python
print(lower_limit)
print(uper_limit)
```
