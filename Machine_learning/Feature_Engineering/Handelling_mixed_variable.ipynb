```python
df["number_numerical"]=pd.to_numeric(df["number"],errors="coerce",downcast="integer")
```

```python
df
```

```python
df["number_categorical"]=np.where(df["number"].isnull(),df["number"],np.nan)
```

```python
df.sample(5)
```

```python
df["Cabin_cat"]=df["Cabin"].str[0]
```

```python
df["Cabin_num"]=df["Cabin"].str.extract("(\d+)")
```

```python
df.sample(5)
```

```python
df["Cabin_cat"].value_counts().plot.bar()
```

```python
df["Ticket_num"]=pd.to_numeric(df["Ticket"].apply(lambda x:x.split()[-1]),errors="coerce",downcast="integer")
```

```python
df["Ticket_cat"]=np.where(df["Ticket"].apply(lambda x:x.split()[0]).str.isdigit(),np.nan,df["Ticket"].apply(lambda x:x.split()[0]))
```
