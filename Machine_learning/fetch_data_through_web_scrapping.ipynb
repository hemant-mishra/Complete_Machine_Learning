```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Set up Chrome
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
# options.add_argument('--headless')  # Optional: run in background

driver = webdriver.Chrome(options=options)

# Open Naukri job page
driver.get("https://www.naukri.com/data-scientist-jobs")

# Wait for job cards to load
time.sleep(5)  # increase if needed

# Grab job titles
titles = driver.find_elements(By.CLASS_NAME, "title")

for title in titles:
    print(title.text)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from time import sleep

driver = webdriver.Chrome()
query="laptop"
file=0
for i in range(1,10):
    driver.get(f"https://www.amazon.in/s?k={query}&page={i}&crid=1VN44CA19GTST&sprefix=laptop%2Caps%2C298&ref=nb_sb_noss_2")
    elems = driver.find_elements(By.CLASS_NAME, "puis-card-container")
    print(elems)
    for elem in elems:
        # print(ele.text)
        d=elem.get_attribute("outerHTML")
        with open(f"{query}_{file}.html","w",encoding="utf-8") as f:
            # print(ele.get_attribute("outerHTML"))
            f.write(d)
            file +=1
    sleep(5)
driver.close()
```

```python
from bs4 import BeautifulSoup
import os
import pandas as pd

list_file=os.listdir()
d={"Title":[],"Price":[]}
for file in list_file:
    try:
        with open(f"{file}",encoding="utf-8") as f:
            html_doc=f.read()
        soup=BeautifulSoup(html_doc,"html.parser")

        t=soup.find("h2")
        title=t.get_text()
        # print(title)

        p=soup.find("span",attrs={"class":"a-price-whole"}).get_text()
        # print(p.get_text())
        d["Price"].append(p)
        d["Title"].append(title.strip())
        # l=t.find("a")
        # link="https://amazon.in/"+ l["href"]
        # print(link)
        # print(soup.prettify())
        # break
    except Exception as e:
        print(e)
df=pd.DataFrame(d)
print(df)
df.to_csv("amazon_latop.csv")
```
