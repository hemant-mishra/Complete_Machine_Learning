```python
import random
class MiniBSGD:
    
    def __init__(self,lr=0.01,epochs=100,batch=50):
        self.lr=lr
        self.epochs=epochs
        self.intercept_=None
        self.coef_=None
        self.batch=batch
    
    def fit(self,x_train,y_train):
        
        self.intercept_=0
        self.coef_=np.ones(x_train.shape[1])
        
        for i in range(self.epochs):
            for j in range(int(x_train.shape[0]/self.batch)):
                idx= random.sample(range(x_train.shape[0]),self.batch)
                
                y_hat=np.dot(x_train[idx],self.coef_) + self.intercept_
                
                slope_intercept_ = -2*np.mean(y_train[idx] - y_hat)
                self.intercept_=self.intercept_ - self.lr*slope_intercept_
                
                slope_coef_= -2*np.dot((y_train[idx]-y_hat),x_train[idx])
                self.coef_ = self.coef_ - self.lr*slope_coef_
        
        print(self.intercept_,self.coef_)
    
    def predict(self,x_test):
        return np.dot(self.coef_,x_test) + self.intercept_
```

```python
mn=MiniBSGD()
```

```python
mn.fit(x_train,y_train)
```
